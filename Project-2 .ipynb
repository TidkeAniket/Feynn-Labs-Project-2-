Based on the provided data, which represents the number of seats in vehicles mostly used between 2010 and 2020, we can observe the following summary:

This summary provides an overview of the distribution of seat counts in vehicles primarily used between 2010 and 2020. The majority of vehicles had 5 or 7 seats, while other seat counts were less prevalent.

```python
#Seats distribution


plt.figure(figsize=(10, 8))
ax = sns.countplot(x='seats', data=df)
plt.title('Count of Cars by Number of Seats')
plt.xlabel('Number of Seats')
plt.ylabel('Count')

# Adding count labels on top of each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')

plt.show()
```

```python
#Finding Average Power no of seat wise
average_power_by_seats = df.groupby('seats')['max_power(bhp)'].mean()
print(average_power_by_seats)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=average_power_by_seats.index, y=average_power_by_seats.values)
plt.title('Average Power by Seats')
plt.xlabel('Seats')
plt.ylabel('Average Power (bhp)')
plt.xticks(rotation=0)
plt.show()
```

```python
#Finding Average engine capacity no of seat wise
average_engine_by_seats = df.groupby('seats')['engine(CC)'].mean()
print(average_engine_by_seats)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=average_engine_by_seats.index, y=average_engine_by_seats.values)
plt.title('Average Engine Size by Seats')
plt.xlabel('Seats')
plt.ylabel('Average Engine Size (CC)')
plt.xticks(rotation=0)
plt.show()
```

```python
#Finding Average mileage of vehicles no of seat wise
average_mileage_by_seats = df.groupby('seats')['mileage(kmpl)'].mean()
print(average_mileage_by_seats)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=average_mileage_by_seats.index, y=average_mileage_by_seats.values)
plt.title('Average Mileage by Seats')
plt.xlabel('Seats')
plt.ylabel('Average Mileage (kmpl)')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='transmission', data=df)
plt.title('Transmission Type Counts')
plt.xlabel('Transmission Type')
plt.ylabel('Count')

# Add count labels on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
# Filter the data for automatic transmissions and from year 2001 onwards
filtered_data = df[(df['transmission'] == 'Automatic') & (df['year'] >= 2001)]

plt.figure(figsize=(12, 6))
sns.countplot(x='year', hue='seats', data=filtered_data)
plt.title('Number of Cars by Year and Seats (Automatic Transmission)')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Number of Seats')
plt.xticks(rotation=90)
plt.show()
```

Importing file of state/UT wise operational Electric vehicle charging stations

```python
data1=pd.read_csv('StateUTs-wise Operational Charging Stations.csv')

data1.head()
```

```python
data1.info()
```

```python
# Extracting data from DataFrame
states = data1['State/UT']
charging_stations = data1['Operational Charging Stations']

plt.figure(figsize=(15,8))

# Plotting the bar chart using sns.barplot
sns.barplot(x=states, y=charging_stations)

# Adding labels and title
plt.xlabel('State/UT')
plt.ylabel('Number of Operational Charging Stations')
plt.title('Operational Charging Stations by State/UT')

# Rotating x-axis labels 
plt.xticks(rotation=90)

# Adding numbers at the top of each bar
for i, value in enumerate(charging_stations):
    plt.text(i, value, str(value), ha='center', va='bottom')

# Display the plot
plt.show()
```
